```python
#importin
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importin data
spikes_left_trial1 = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/spikes_left_trial1.csv', delimiter = ',')
spikes_left_trial2 = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/spikes_left_trial2.csv', delimiter = ',')
spikes_right_trial1 = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/spikes_right_trial1.csv', delimiter = ',')
spikes_right_trial2 = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/spikes_right_trial2.csv', delimiter = ',')
```

```python
#printing length of arrays
print(len(spikes_left_trial1),
      len(spikes_left_trial2),
      len(spikes_right_trial1),
      len(spikes_right_trial2))
```

```python
#determine firing rate
left1fr = len(spikes_left_trial1)/30
left2fr = len(spikes_left_trial2)/30
right1fr = len(spikes_right_trial1)/30
right2fr = len(spikes_right_trial2)/30
print(left1fr, left2fr, right1fr, right2fr)
```

The differences in firing rate are pretty odd. The left side has a higher firing rate on the first trial, and it decreases after that trial. The right side is lower on the first trial, but increases in the second to be the higher.

```python
#plotting Spiketrains
fig, ax = plt.subplots(2)
ax[0].plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.')
ax[0].plot(spikes_right_trial1, 2 * np.ones_like(spikes_right_trial1), '.')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylim([0,3])
ax[0].set_yticks([])
ax[1].plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.')
ax[1].plot(spikes_right_trial2,2 * np.ones_like(spikes_right_trial2), '.')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylim([0,3])
plt.yticks([])
fig.tight_layout()
```

In general the neurons seem to be pretty equally bursty. In the second trial the left neurons seem to fire more and might be slightly more bursty, but the right neurons are still clumped in tight bursts as well. Both neurons seem to be very similar. Which makes sense considering they perform the same function.

```python
#calculating ISI's
ISI_Left1 = np.diff(spikes_left_trial1)
ISI_Left2 = np.diff(spikes_left_trial2)
ISI_Right1 = np.diff(spikes_right_trial1)
ISI_Right2 = np.diff(spikes_right_trial2)
```

```python
#plotting the ISI's in histograms to compare
fig, ax = plt.subplots(2,2)
bins = np.arange(0, 2, .01)
ax[0,0].hist(ISI_Left1, bins)
ax[0,0].set_xlabel('ISI [s]')
ax[0,0].set_ylabel('Counts')
ax[0,0].set_title('Left Trial 1')
ax[1,0].hist(ISI_Left2, bins)
ax[1,0].set_xlabel('ISI [s]')
ax[1,0].set_ylabel('Counts')
ax[1,0].set_title('Left Trial 2')
ax[0,1].hist(ISI_Right1, bins)
ax[0,1].set_xlabel('ISI [s]')
ax[0,1].set_ylabel('Counts')
ax[0,1].set_title('Right Trial 1')
ax[1,1].hist(ISI_Right2, bins)
ax[1,1].set_xlabel('ISI [s]')
ax[1,1].set_ylabel('Counts')
ax[1,1].set_title('Right Trial 2')
fig.tight_layout()
```

Both neurons seem to be pretty equally burtsy, with the left trial seeming to have a slightly higher density of low ISI's in trial 1, but with the right neurons having the higher concentration of low ISI values in trial 2.

```python
#calculating firing rate across time and plotting, defining a funtion that calculates firing rate as to not have to repeat the for loop.
fig, ax = plt.subplots(2)
binsize = 1
fr_bins = np.arange(0,30,binsize)
fr_low = np.zeros(len(fr_bins)-1)
def firingrate(x):
  for i in range(len(fr_bins)-1):
    spk_index = (x > fr_bins[i]) & (x < fr_bins[i+1])
    true_values = x[spk_index]
    fr_low[i] = (len(true_values))/binsize
firingrate(spikes_left_trial1)
ax[0].plot(fr_bins[:-1], fr_low, color = 'r')
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Firing Rate [Hz]')
ax[0].set_title('Left Trials')
firingrate(spikes_left_trial2)
ax[0].plot(fr_bins[:-1], fr_low, color = 'b')
firingrate(spikes_right_trial1)
ax[1].plot(fr_bins[:-1], fr_low, color = 'r')
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Firing Rate [Hz]')
ax[1].set_title('Right Trials')
firingrate(spikes_right_trial2)
ax[1].plot(fr_bins[:-1], fr_low, color = 'b')
fig.tight_layout()
```

Left trials seem to have a higher firing rate in the middle of the trial while the right trials seem to have a higher firing rate at the beginning of the trials. Both neurons have large differences between trials, with left spiking at the midpoint in trial 1 and towards the end in trial 2. And the right side neurons has a lower firing rate throughout trial 1 spiking a little in the middle and firing much faster in the beginning in trial 2.
