```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#importing data
v_rest_sst = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/v_rest_sst.csv')
v_rest_pvalb = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/v_rest_pvalb.csv')

```

```python
print(np.mean(v_rest_sst)) #finding means
print(np.mean(v_rest_pvalb))

t, pvalue = stats.ttest_ind(v_rest_sst,v_rest_pvalb) #performing ttest, seperating values
print(stats.ttest_ind(v_rest_sst,v_rest_pvalb))

if pvalue > .05: #creating the condition of figure being produced if p value is less than .05
  fig, ax = plt.subplots(1,2,)
  ax[0].hist(v_rest_sst)
  ax[1].hist(v_rest_pvalb)
  ax[0].set_xlabel('Resting Membrane Potential')
  ax[0].set_ylabel('Counts')
  ax[1].set_xlabel('Resting Membrane Potential')
  ax[1].set_ylabel('Counts')
else:
  print('womp womp') #I know this wasn't in the instructions but I thought it was funny

```

```python
def vresthist(): #creates a function that will run all of the code from the last cell
  print(np.mean(v_rest_sst))
  print(np.mean(v_rest_pvalb))

  t, pvalue = stats.ttest_ind(v_rest_sst,v_rest_pvalb)
  print(stats.ttest_ind(v_rest_sst,v_rest_pvalb))

  if pvalue > .05:
    fig, ax = plt.subplots(1,2,)
    ax[0].hist(v_rest_sst)
    ax[1].hist(v_rest_pvalb)
    ax[0].set_xlabel('Resting Membrane Potential')
    ax[0].set_ylabel('Counts')
    ax[1].set_xlabel('Resting Membrane Potential')
    ax[1].set_ylabel('Counts')
  else:
    print('womp womp')
```

```python
vresthist() #running the function
```

```python
#creating a loop that looks into each row of the two lists, and changes the color based on their values.


for i in range(len(v_rest_sst)):
    sst = v_rest_sst[i] #assigning each to a variable to make rest easier
    pvalb = v_rest_pvalb[i]

    if sst > -60 and pvalb > -60:
        color = 'red'  #Plotting red if both are over -60
    elif sst > -60 and pvalb <= -60:
        color = 'green'  #plotting green if sst >-60, but Pvalb <-60
    elif sst < -60 and pvalb < -60:
        color = 'blue'  #plotting blue if both values are less than -60
    elif sst < -60 and pvalb > -60:
        color = 'purple' #plotting purple if sst <-60, but pvalb >-60

    plt.scatter(sst, pvalb, color=color)

plt.xlabel('SST Resting Membrane Potential')
plt.ylabel('Pvalb Resting Membrane Potential')

```
