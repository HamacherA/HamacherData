```python
#importin stuff
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data
acc_fa = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/acc_fa_hw6.csv', delimiter = ',')
acc_hits = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/acc_hits_hw6.csv', delimiter = ',')
lc_hits = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/lc_hits_hw6.csv', delimiter = ',')
lc_fa = np.loadtxt('https://raw.githubusercontent.com/HamacherA/HamacherData/main/lc_fa_hw6.csv', delimiter = ',')
```

```python
#printing the shape of each array
print(
    np.shape(acc_fa),
    np.shape(acc_hits),
    np.shape(lc_hits),
    np.shape(lc_fa)
)
```

```python
#creating time variable
time = np.linspace(0,4,8001)
```

```python
#determining sampling rate
len(time)/time[-1]
```

```python
#plotting the first trial of both hits and FA's for both sets of data
fig, ax = plt.subplots(2)
ax[0].plot(time, acc_hits[0], color = 'r')
ax[0].plot(time,lc_hits[0], color = 'b')
ax[0].set_xlabel('time(s)')
ax[0].set_ylabel('LFP($\mu v$)')
ax[1].plot(time, acc_fa[0], color = 'r')
ax[1].plot(time,lc_fa[0], color = 'b')
ax[1].set_xlabel('time(s)')
ax[1].set_ylabel('LFP($\mu v$)')

fig.tight_layout()
```

While not the clearest LFP's looking at the hits graph makes me think that the ACC has a dominant oscilaiton with very low frequency, something in the 1-3Hz range. The LC on the other hand has a very high frequency, to the point where it is too high to count manually, probably in the 40-70Hz range. There does seem to be some phase amplitude coupling occuring, with the LC having greater amplitudes on the downward phase of the ACC frequency, specifically on the later half of the sample. There may be some small amount of coherence, as around the 1 sec mark in the hits trials the peaks lineup quite nicely.

```python
# Following code from coherence notebook on computing coherence (making it a function so stop 12 is easier)
def hits_coherence():
  xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
  yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

  #set the sampling interval and number of points per trial
  dt = time[1] - time[0]
  N = acc_hits.shape[1]


  T = time[-1]
  scale = 2 * dt**2 / T
  Sxx = scale * (xf * xf.conj()).mean(0)
  Syy = scale * (yf * yf.conj()).mean(0)
  Sxy = scale * (xf * yf.conj()).mean(0)

  # Compute the coherence.
  cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

  f = np.fft.rfftfreq(N, dt)
  plt.plot(f, cohr.real, color= 'r')
  plt.xlim([0, 50])
  plt.ylim([0, 1])
  plt.xlabel('Frequency [Hz]')
  plt.ylabel('Coherence')
  plt.title('Coherence between Acc_Hits and LC_Hits')
hits_coherence()
```

There seems to be an incredibly high coherence between the ACC and LC in the hits trials. Some notable spikes occur in the 1-3Hz ranges and 5-10 range, as well as a large spike around 38Hz, as well as around 25Hz.

```python
# Following code from coherence notebook on computing coherence for FA's
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real, color = 'b')
plt.xlim([0, 50])
plt.ylim([0, 1])

hits_coherence()
plt.title('Coherence for Both')
```

The coherence seems to be very similar between the two, with the main difference being hits having a much higher coherence around 38 hz as well as around 25Hz. in general, most of the spikes in the later range (>25Hz) seem to be stronger.
